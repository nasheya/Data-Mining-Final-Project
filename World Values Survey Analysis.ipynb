```python
import statsmodels.api as sm

x = [hpop_hdi["V15_1"], hpop_hdi["V16_1"], hpop_hdi["V20_1"], hpop_hdi["V22_1"], hpop_hdi["V27_1"], hpop_hdi["V32_1"], hpop_hdi["V27_2"], hpop_hdi["V32_2"], hpop_hdi["V70_1"], hpop_hdi["V70_2"], hpop_hdi["V70_3"], hpop_hdi["V74B_1"], hpop_hdi["V74B_2"], hpop_hdi["V74B_3"], hpop_hdi["V143_1"], hpop_hdi["V143_2"], hpop_hdi["V160E_1"], hpop_hdi["V160E_2"], hpop_hdi["V160J_5"],hpop_hdi["V160J_4"]]

def reg_m(y, x):
    ones = np.ones(len(x[0]))
    X = sm.add_constant(np.column_stack((x[0], ones)))
    for ele in x[1:]:
        X = sm.add_constant(np.column_stack((ele, X)))
    results = sm.OLS(y, X).fit()
    return results

print reg_m(hpop_hdi["Human Development Index (HDI)"], x).summary()
```
